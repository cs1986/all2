```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df= sns.load_dataset('titanic')
df=df.dropna()

sns.displot(df["fare"],bins=10)
sns.jointplot(y='age',x='fare',data=df,kind='hex')
sns.rugplot(df['fare'])
sns.pairplot(df,hue='sex')
```

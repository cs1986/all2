```python
import pandas as pd 
import numpy as np
from sklearn import datasets
```

```python
url = "income.csv"
df = pd.read_csv(url)
df.head()
```

```python
print(df.age_group.unique())
print(df.income.unique())
```

```python
df.groupby(df.age_group).count()
```

```python
df.groupby(df.age_group).min()
```

```python
df.groupby(df.age_group).max()
```
